visualization task

bar chart

```python
plt.figure(figsize=(10,6))
avg_scores_by_gender = df.groupby('gender')[['math score', 'reading score', 'writing score']].mean()
avg_scores_by_gender.plot(kind='bar')
plt.title('Average Scores by Gender')
plt.ylabel('Average Score')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

Pie Chart

```python
plt.figure(figsize=(8,8))
parental_education = df['parental level of education'].value_counts()
plt.pie(parental_education, labels=parental_education.index, autopct='%1.1f%%', startangle=140)
plt.title('Parental Level of Education Distribution')
plt.show()
```

1. Bar Chart - Average Scores by Gender:
This bar chart helps compare average performance between male and female students across math, reading, and writing. It makes it easier to identify gender-based disparities or similarities in subject performance.

2. Pie Chart - Parental Level of Education:
This pie chart illustrates the proportion of students based on their parents' education levels. It provides insight into the diversity of educational backgrounds among students' families.

3. Box Plot - Score Distribution for Each Subject:
The box plot shows the distribution and spread of scores, highlighting medians and potential outliers for each subject. It allows quick identification of variability and score ranges in math, reading, and writing.

4. Scatter Plot - Math Score vs. Writing Score:
The scatter plot shows the relationship between math and writing scores for individual students. It helps identify any correlation or patterns in performance between these two subjects.

5. Heatmap - Correlation Between Variables:
This heatmap visualizes the strength and direction of correlations between math, reading, and writing scores. It makes it easier to understand how performance in one subject is related to performance in others.
